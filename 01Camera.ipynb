```python
%%capture_png -p Template.png --celltype Header --style "width:120px"
import matplotlib.pyplot as plt
import napari
from skimage import data
                                                                             # NOT
from napari.settings import get_settings                                     # NOT                                        
                                                                             # NOT     
get_settings().application.window_position = (1000, 300)                     # NOT    
get_settings().application.window_fullscreen = False                         # NOT                                                                                                                      
get_settings().application.window_statusbar = False                          # NOT                                                   
get_settings().application.save_window_state = False                         # NOT                                                    
viewer = napari.Viewer()
viewer.theme = "light"
viewer.dims.ndisplay = 3
viewer.window.resize(800, 600)                                               # NOT                                                            
viewer.window.qt_viewer.dockLayerControls.toggleViewAction().trigger()       # NOT                                                                                                    
viewer.window.qt_viewer.dockLayerList.toggleViewAction().trigger()           # NOT                                                                                                
                                                                             # NOT                                                                                                                      
                                                                             # NOT                                
def make_screenshot(viewer=None, region=None):                               # NOT                               
    img = viewer.screenshot(canvas_only=True, flash=False)                   # NOT                                           
    if region:                                                               # NOT                                              
        print(img.shape)                                                     # NOT         
        img = img[region[0] : region[1], region[2] : region[3]]              # NOT                                                
    plt.imshow(img)                                                          # NOT
    plt.axis("off")                                                          # NOT
                                                                             # NOT   
                                                                             # NOT                                                                 
# make data                                                                  # NOT        
blobs = data.binary_blobs(length=301, volume_fraction=0.1, n_dim=3, seed=42)
blobs = blobs[200:, 200:, :-100]
blobs[:, :, :50] = 1
                                                                              # NOT
import matplotlib.pyplot as plt                                               # NOT
                                                                              # NOT
%config InlineBackend.print_figure_kwargs = {'pad_inches':0}                  # NOT
                                                                              # NOT
                                                                              # NOT
fix, ax = plt.subplots(figsize=(5, 4), facecolor="#F7F5F3")                   # NOT
text_kwargs = dict(ha="center", va="center", fontsize=70, color="#ED6A5A")    # NOT
ax.text(0.5, 0.5, "Header", **text_kwargs)                                    # NOT
ax.axis("off");                                                               # NOT
```

```python
%%capture_png -p FirstView.png --celltype Normal --style "width:120px"
viewer.reset_view()
viewer.layers.clear()
viewer.dims.ndisplay = 3
layer1 = viewer.add_labels(blobs, name="layer1", seed=0.03)
layer1.opacity = 1
make_screenshot(viewer)  # NOT
```
