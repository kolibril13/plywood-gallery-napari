```python
%%capture_png -p FirstView.png --celltype Header --style "height:200px; width:auto;"

import napari
import matplotlib.pyplot as plt
from skimage import data

%config InlineBackend.print_figure_kwargs = {'pad_inches':0} # NOT


def make_screenshot(viewer):
    img = viewer.screenshot(canvas_only=False, flash=False)
    plt.imshow(img)
    plt.axis("off")


# make data
blobs = data.binary_blobs(length=301, volume_fraction=0.1, n_dim=3, seed=42)
blobs = blobs[200:, 200:, :-100]
blobs[:, :, :50] = 1

viewer = napari.Viewer()
viewer.theme = "dark"
viewer.dims.ndisplay = 3
layer1 = viewer.add_labels(blobs, name="layer1", seed=0.03)
viewer.camera.angles = (30, 40, 80)
layer1.opacity = 1
make_screenshot(viewer)
```

```python
%%capture_png -p NewTheme.png --celltype Normal --style "height:200px; width:auto;"
viewer.theme = "light"
make_screenshot(viewer)
```
